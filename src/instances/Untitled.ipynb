```python
from matplotlib import pyplot as plt
```

```python
values = {'instance01':[], 'instance02':[], 'instance03':[], 'instance04':[], 'instance05':[], 'instance06':[], 'instance07':[], 'instance08':[]}

with open("perFilippo.csv", 'r') as f:
    for line in f.readlines():
        if len(line) > 0:
            field = line.split()
            values[field[0]].append(float(field[2]))
print(values)
        
    
```

```python
plt.boxplot(values["instance01"])
plt.title("instance01")
plt.savefig("instance01.png")

```
